### Import Libraries

```python
import pandas as pd
import pyodbc 
from textblob import TextBlob
```

### Initiate Connection with Database

```python
server = 'DESKTOP-6POBDOL'
database = 'ShopEasy'

conn = pyodbc.connect(
    f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'
)
```

### Query customer_reviews Table and Store it in a Dataframe

```python
query = 'SELECT * FROM dbo.customer_reviews'
df = pd.read_sql(query , conn)
```

### Display Data

```python
df.head()
```

### Divide Reviews into three Groups (Positive , Negative , Neutral)

```python
def analyze_sentiment(text):
    blob = TextBlob(str(text))
    polarity = blob.sentiment.polarity  # a value from -1 to +1

    if polarity > 0.15:
        sentiment = 'Positive'
    elif polarity < -0.15:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'

    return pd.Series([polarity, sentiment])
```

### Apply function to our Dataframe

```python
df[['Polarity_Score', 'Sentiment']] = df['ReviewText'].apply(analyze_sentiment)
```
